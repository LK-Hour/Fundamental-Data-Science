```python
import pandas as pd

file_path = '/content/drive/MyDrive/Data Science/Smartphone_Usage_Productivity_Dataset_50000.csv'
df = pd.read_csv(file_path)

print(f"Successfully loaded data from: {file_path}")
display(df.head())
```

```python
# Exercise 1
# How many rows and columns does your data have?
# What dose each column represent?

# Get the number of rows and columns
rows, columns = df.shape
print(f"The dataset has {rows} rows and {columns} columns.\n")

# Get information about each column (name, non-null count, and data type)
print("Column Information and Representation:\n")
df.info()
```

```python
# Exercise 2
# Which columns are categorical (nominal or ordinal) variables?
# Which columns are numerical (interval or ratio) variables?

# Identify numerical columns
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
print("Numerical Variables:")
for col in numerical_cols:
    print(f"- {col}")

print("\n")

# Identify categorical columns
categorical_cols = df.select_dtypes(include=['object']).columns.tolist()
print("Categorical Variables:")
for col in categorical_cols:
    print(f"- {col}")
```

```python
# Exercise 3
# Calculate Minimum of numerical variable and explain the result

print("Minimum values for numerical variables:")
for col in numerical_cols:
    min_val = df[col].min()
    print(f"- {col}: {min_val}")

print("\nExplanation:\nThe minimum value for each numerical variable represents the lowest observed data point in that particular column. For example, the minimum 'Age' indicates the youngest person in the dataset, and the minimum 'Daily_Phone_Hours' indicates the least amount of daily phone usage recorded.")
```

```python
# Exercise 4
# Calculate Maximum of numerical variable and explain the result

print("Maximum values for numerical variables:")
for col in numerical_cols:
    max_val = df[col].max()
    print(f"- {col}: {max_val}")

print("\nExplanation:\nThe maximum value for each numerical variable represents the highest observed data point in that particular column. For example, the maximum 'Age' indicates the oldest person in the dataset, and the maximum 'Daily_Phone_Hours' indicates the most amount of daily phone usage recorded.")
```

```python
# Exercise 5
# Calculate mean of numerical variable and explain the result

print("Mean values for numerical variables:")
for col in numerical_cols:
    mean_val = df[col].mean()
    print(f"- {col}: {mean_val:.2f}") # Format to 2 decimal places for readability

print("\nExplanation:\nThe mean value for each numerical variable represents the average value of all observed data points in that particular column. It is calculated by summing all values and dividing by the number of observations. For example, the mean 'Age' indicates the average age of individuals in the dataset, and the mean 'Daily_Phone_Hours' indicates the average daily phone usage.")
```

```python
# Exercise 6
# Calculate Median of numerical variable and explain the result

print("Median values for numerical variables:")
for col in numerical_cols:
    median_val = df[col].median()
    print(f"- {col}: {median_val:.2f}") # Format to 2 decimal places for readability

print("\nExplanation:\nThe median value for each numerical variable represents the middle value in a sorted list of all observed data points for that particular column. If there's an odd number of observations, it's the exact middle. If there's an even number, it's the average of the two middle values. The median is less sensitive to extreme outliers than the mean. For example, the median 'Age' indicates the age at which half the individuals are younger and half are older, and the median 'Daily_Phone_Hours' indicates the point where half the users spend less time and half spend more time on their phone daily.")
```

```python
# Exercise 7
# Calculate Mode of numerical variable and explain the result

print("Mode values for numerical variables:")
for col in numerical_cols:
    mode_val = df[col].mode()
    # Mode can return multiple values if there are ties, so we print all of them
    print(f"- {col}: {mode_val.tolist()}")

print("\nExplanation:\nThe mode for each numerical variable represents the value that appears most frequently in that column. Unlike the mean or median, there can be one mode (unimodal), multiple modes (multimodal), or no mode if all values appear with the same frequency. The mode is useful for understanding the most common observation within a dataset. For example, the mode of 'Age' would indicate the most common age among the individuals in the dataset.")
```

```python
# Exercise 8
# Calculate Variance of numerical variable and explain the result

print("Variance values for numerical variables:")
for col in numerical_cols:
    var_val = df[col].var()
    print(f"- {col}: {var_val:.2f}") # Format to 2 decimal places for readability

print("\nExplanation:\nThe variance measures the average squared difference from the mean for each data point in a numerical variable. It gives an idea of the spread of the data. A high variance indicates that data points are widely spread out from the mean, while a low variance suggests that data points are clustered closely around the mean. Unlike standard deviation, variance is in squared units, which can make it less intuitive to interpret directly in the context of the original data units, but it's crucial for many statistical analyses.")
```

```python
# Exercise 9
# Calculate Standard Deviation of numerical variable and explain the result

print("Standard Deviation values for numerical variables:")
for col in numerical_cols:
    std_val = df[col].std()
    print(f"- {col}: {std_val:.2f}") # Format to 2 decimal places for readability

print("\nExplanation:\nThe standard deviation measures the average amount of variability or dispersion in a set of data. A low standard deviation indicates that the data points tend to be close to the mean (also known as the expected value) of the set, while a high standard deviation indicates that the data points are spread out over a wider range of values. For example, a low standard deviation for 'Age' means most ages are close to the average age, whereas a high standard deviation would mean ages are more widely distributed.")
```

```python
# Exercise 10
# Calculate Q1, Q2, Q3 of numerical variable and explain the result

print("Quartile values for numerical variables:")
for col in numerical_cols:
    q1 = df[col].quantile(0.25)
    q2 = df[col].quantile(0.50)
    q3 = df[col].quantile(0.75)
    print(f"- {col}:")
    print(f"    Q1 (25th percentile): {q1:.2f}")
    print(f"    Q2 (50th percentile/Median): {q2:.2f}")
    print(f"    Q3 (75th percentile): {q3:.2f}\n")

print("Explanation:\nQuartiles divide a dataset into four equal parts. Q1 (the first quartile or 25th percentile) means 25% of the data falls below this value. Q2 (the second quartile or 50th percentile) is the median, meaning 50% of the data falls below this value. Q3 (the third quartile or 75th percentile) means 75% of the data falls below this value. These measures help to understand the spread and central tendency of the data, especially useful for identifying skewness and potential outliers.")
```

```python
# Exercise 11
# plot Boxplot and explain result

import matplotlib.pyplot as plt
import seaborn as sns

print("Generating Boxplots for Numerical Variables:")
for col in numerical_cols:
    plt.figure(figsize=(8, 6))
    sns.boxplot(y=df[col])
    plt.title(f'Boxplot of {col}')
    plt.ylabel(col)
    plt.show()

print("""\nExplanation:\nA boxplot, also known as a box-and-whisker plot, displays the distribution of a numerical variable and provides a visual summary of its key statistical measures:
- The central line inside the box represents the median (Q2, 50th percentile).
- The box itself spans from the first quartile (Q1, 25th percentile) to the third quartile (Q3, 75th percentile), covering the interquartile range (IQR).
- The 'whiskers' extend from the box to the minimum and maximum values within 1.5 times the IQR from the quartiles. Data points outside these whiskers are typically considered outliers.

Boxplots are particularly useful for:
1.  **Identifying Central Tendency**: The median line shows where the bulk of the data lies.
2.  **Assessing Spread**: The length of the box (IQR) and whiskers indicates the variability of the data.
3.  **Detecting Skewness**: If the median is not in the middle of the box, or if one whisker is significantly longer than the other, it suggests skewness.
4.  **Highlighting Outliers**: Individual points plotted outside the whiskers are potential outliers, indicating extreme values in the dataset.""")
```

```python
# Exercise 13
# calculate lower_bound of outlier and upper_bound of outlier and explain the result

print("Outlier Bounds for Numerical Variables:")
for col in numerical_cols:
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    print(f"- {col}:")
    print(f"    Lower Bound: {lower_bound:.2f}")
    print(f"    Upper Bound: {upper_bound:.2f}")
    outliers_low = df[df[col] < lower_bound]
    outliers_high = df[df[col] > upper_bound]
    print(f"    Number of potential outliers below lower bound: {len(outliers_low)}")
    print(f"    Number of potential outliers above upper bound: {len(outliers_high)}\n")

print("""Explanation:
The lower and upper bounds for outliers are calculated using the Interquartile Range (IQR) method. The IQR is the range between the first quartile (Q1) and the third quartile (Q3).

-   **Lower Bound**: Q1 - 1.5 * IQR
-   **Upper Bound**: Q3 + 1.5 * IQR

Any data point that falls below the lower bound or above the upper bound is typically considered a potential outlier. These bounds are crucial for identifying extreme values that might warrant further investigation, as they could be errors, anomalies, or genuinely rare occurrences within the dataset. The 1.5 factor is a commonly used heuristic, but it can be adjusted based on the specific domain and data characteristics.""")
```

```python
# Exercise 14
# calculate Coefficient of Variation of numerical variable and explain the result
# When CV should use?

print("Coefficient of Variation (CV) for numerical variables:")
for col in numerical_cols:
    mean_val = df[col].mean()
    std_val = df[col].std()

    if mean_val != 0:
        cv_val = (std_val / mean_val) * 100
        print(f"- {col}: {cv_val:.2f}%")
    else:
        print(f"- {col}: Mean is zero, CV cannot be calculated.")

print("""\nExplanation:
The Coefficient of Variation (CV) is a standardized measure of dispersion of a probability distribution or frequency distribution. It is often expressed as a percentage and is calculated as the ratio of the standard deviation to the mean:

CV = (Standard Deviation / Mean) * 100%

Key aspects and when to use CV:
1.  **Relative Variability**: Unlike standard deviation, which measures absolute variability, CV expresses variability relative to the mean. This makes it useful for comparing the extent of variation between datasets with different units, or vastly different means.
2.  **Comparison Across Different Scales**: For example, comparing the variability of daily phone hours (mean ~6.5 hours) with App Usage Count (mean ~32 counts). Directly comparing their standard deviations might be misleading due to the difference in scale and units. CV provides a scale-free measure for such comparisons.
3.  **Risk Assessment**: In finance, CV is often used to compare the risk (standard deviation) of investments with different expected returns (mean). A lower CV indicates a more consistent return relative to its expected value.
4.  **Quality Control**: In manufacturing or scientific experiments, CV can assess the precision and reliability of measurements or processes. A low CV indicates high precision.

In summary, use CV when you need to compare the relative dispersion of data from two or more series with different units or means, to understand which dataset has more variability *relative to its average*.""")
```

```python
# Exercise 15
# Calculate covariance of two numerical variable and explain the result

# Choosing two numerical variables for demonstration
var1 = 'Daily_Phone_Hours'
var2 = 'Social_Media_Hours'

covariance_val = df[var1].cov(df[var2])

print(f"Covariance between {var1} and {var2}: {covariance_val:.2f}")

print("""\nExplanation:\nCovariance is a measure of how two random variables change together. It indicates the direction of the linear relationship between two variables. A positive covariance suggests that the variables tend to move in the same direction (as one increases, the other tends to increase). A negative covariance suggests that they tend to move in opposite directions (as one increases, the other tends to decrease). A covariance close to zero suggests no linear relationship between the variables. \n\nHowever, the magnitude of covariance is difficult to interpret because it is not standardized; it depends on the units of the variables. For example, if we changed 'Daily_Phone_Hours' to 'Daily_Phone_Minutes', the covariance value would change significantly, even if the underlying relationship remained the same. For a standardized measure of the strength and direction of a linear relationship, correlation is generally preferred (see next exercise).""")
```

```python
# Exercise 16
# Calculate Correlation of two numerical variable and explain the result
# What is the difference between correlation and covariance?

# Choosing the same two numerical variables for consistency
var1 = 'Daily_Phone_Hours'
var2 = 'Social_Media_Hours'

correlation_val = df[var1].corr(df[var2])

print(f"Correlation between {var1} and {var2}: {correlation_val:.2f}")

print("""\nExplanation of Correlation:\nCorrelation is a statistical measure that expresses the extent to which two variables are linearly related (meaning they change together at a constant rate). It's a standardized measure, ranging from -1 to +1.

-   **+1 (Positive Correlation)**: Indicates a perfect positive linear relationship. As one variable increases, the other also increases proportionally.
-   **-1 (Negative Correlation)**: Indicates a perfect negative linear relationship. As one variable increases, the other decreases proportionally.
-   **0 (No Linear Correlation)**: Indicates no linear relationship between the two variables. Changes in one variable do not predict changes in the other.

Explanation of the Difference between Correlation and Covariance:

**Covariance**:
-   Measures the direction of the linear relationship between two variables (positive, negative, or none).
-   Its magnitude is not standardized, meaning it depends on the units of the variables. Therefore, it's difficult to interpret the strength of the relationship directly from its value.
-   Range: (-infinity, +infinity)

**Correlation (specifically, Pearson correlation coefficient)**:
-   Measures both the direction and the strength of the linear relationship between two variables.
-   It is a standardized measure, making its magnitude easy to interpret regardless of the variables' units.
-   Range: [-1, +1]

In essence, correlation is a normalized version of covariance, providing a more interpretable and comparable measure of the linear relationship between two variables. If you divide covariance by the product of the standard deviations of the two variables, you get the correlation coefficient.""")
```

```python
# Exercise 17
# calculate Skewness of numerical variable and explain the result

print("Skewness values for numerical variables:")
for col in numerical_cols:
    skew_val = df[col].skew()
    print(f"- {col}: {skew_val:.2f}") # Format to 2 decimal places for readability

print("""\nExplanation:
Skewness is a measure of the asymmetry of the probability distribution of a real-valued random variable about its mean. In simpler terms, it tells us about the shape of the distribution, specifically whether the data is concentrated on one side or the other.

-   **Positive Skew (Right-Skewed)**: If the skewness value is positive, the tail of the distribution is longer on the right side. This means there are more observations with higher values, pulling the mean to the right of the median. In a right-skewed distribution, typically Mean > Median > Mode.
-   **Negative Skew (Left-Skewed)**: If the skewness value is negative, the tail of the distribution is longer on the left side. This means there are more observations with lower values, pulling the mean to the left of the median. In a left-skewed distribution, typically Mean < Median < Mode.
-   **Zero Skew (Symmetric)**: If the skewness value is close to zero, the distribution is roughly symmetric. The mean, median, and mode are approximately equal.

Understanding skewness is important because many statistical models assume a normal (symmetric) distribution. High skewness might indicate the need for data transformation before applying such models or could point to interesting characteristics of the underlying data process.""")
```

```python
# Exercise 18
# Plot Distribution of numerical variable and explain the result

import matplotlib.pyplot as plt
import seaborn as sns

print("Generating Distribution Plots for Numerical Variables:")
for col in numerical_cols:
    plt.figure(figsize=(10, 6))
    sns.histplot(df[col], kde=True, bins=30) # Histogram with KDE overlay
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency / Density')
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.show()

print("""\nExplanation:
Distribution plots, such as histograms combined with Kernel Density Estimates (KDEs), visually represent the shape of a numerical variable's data distribution. They help in understanding:
1.  **Shape of the distribution**: Whether it's symmetric (bell-shaped), skewed (left or right), bimodal, or uniform.
2.  **Central Tendency**: Where the majority of the data is concentrated (e.g., around the mean or median).
3.  **Spread/Variability**: How dispersed the data points are.
4.  **Outliers**: Extreme values that fall far from the main body of the data.

-   **Histograms**: Use bars to show the frequency or count of data points falling into specified ranges (bins).
-   **Kernel Density Estimate (KDE)**: A smoothed, continuous curve that estimates the probability density function of the variable. It helps to visualize the overall shape of the distribution more clearly, especially for identifying peaks and tails.

By examining these plots, we can gain insights into the typical range of values, the presence of multiple peaks, and any significant deviations from a normal distribution.""")
```

```python
# Exercise 19
# Calculate Kurtosis  of numerical variable and explain the result

print("Kurtosis values for numerical variables:")
for col in numerical_cols:
    kurt_val = df[col].kurtosis()
    print(f"- {col}: {kurt_val:.2f}") # Format to 2 decimal places for readability

print("""\nExplanation:
Kurtosis is a statistical measure that describes the 'tailedness' of a distribution. In simpler terms, it tells us how many outliers there are and how extreme they are, which influences the shape of the tails and the peakedness of the distribution compared to a normal distribution.

-   **Mesokurtic (Kurtosis ~ 0)**: Distributions with kurtosis near 0 (like a normal distribution) have tails similar to those of a normal distribution. They are considered to have a moderate number of outliers.
-   **Leptokurtic (Kurtosis > 0)**: Distributions with positive kurtosis have heavier tails and a sharper peak than a normal distribution. This indicates that the data has more extreme outliers than a normal distribution.
-   **Platykurtic (Kurtosis < 0)**: Distributions with negative kurtosis have lighter tails and a flatter peak than a normal distribution. This suggests that the data has fewer and less extreme outliers than a normal distribution.

Understanding kurtosis helps in assessing the risk associated with extreme values and whether a distribution is suitable for statistical models that assume normality. A high kurtosis often implies a higher chance of extreme events.""")
```

```python
# Exercise 20
# calculate Z-score of numerical variable and explain the result

print("Calculating Z-scores for numerical variables:")

z_scores = df[numerical_cols].apply(lambda x: (x - x.mean()) / x.std())

print("First 5 rows of Z-scores for numerical variables:")
display(z_scores.head())

print("""\nExplanation:
A Z-score (also called a standard score) is a numerical measurement that describes a value's relationship to the mean of a group of values, measured in terms of standard deviations from the mean. If a Z-score is 0, it indicates that the data point's score is identical to the mean score.

-   A **positive Z-score** indicates the data point is above the mean.
-   A **negative Z-score** indicates the data point is below the mean.

The magnitude of the Z-score indicates how far the data point is from the mean. For example, a Z-score of 1 means the data point is one standard deviation above the mean, while a Z-score of -2 means it's two standard deviations below the mean.

Z-scores are useful for:
1.  **Standardizing data**: They allow for comparison of observations from different normal distributions.
2.  **Identifying outliers**: Data points with Z-scores typically greater than +3 or less than -3 are often considered outliers.
3.  **Understanding relative position**: They show where a data point stands relative to the rest of the dataset.""")
```

```python
# Exercise 21
# Identify outlier based on Z-score and print the result

# Define a Z-score threshold for outliers (commonly +/- 3)
z_score_threshold = 3

# Identify rows where any numerical variable has an absolute Z-score greater than the threshold
outliers_zscore = df[(z_scores.abs() > z_score_threshold).any(axis=1)]

print(f"Number of potential outliers identified by Z-score (absolute Z > {z_score_threshold}): {len(outliers_zscore)}")

if not outliers_zscore.empty:
    print("\nFirst 5 potential outliers (based on Z-score):")
    display(outliers_zscore.head())
else:
    print("\nNo potential outliers identified based on the Z-score threshold of +/- 3.")

print("""\nExplanation:
Outliers are often identified using Z-scores by setting a threshold. A common rule of thumb is that any data point with an absolute Z-score greater than 2 or 3 is considered an outlier.

-   If |Z-score| > 3 (or sometimes > 2), the data point is significantly far from the mean and is flagged as a potential outlier.
-   A Z-score indicates how many standard deviations an observation is away from the mean. Therefore, a large absolute Z-score means the observation is unusually far from the average.

In this exercise, we used a threshold of +/- 3. If any row contains at least one numerical variable whose Z-score falls outside this range, that entire row is considered an outlier. The presence or absence of outliers is important for understanding data quality and potential impacts on statistical models.""")
```
